```python
from google.colab import userdata
from transformers import pipeline
import gradio as gr

hf_token = userdata.get('HF_TOKEN')
```

```python
#summarize = pipeline("summarization")
#summarize = pipeline("summarization", model="openai/gpt-oss-safeguard-20b")
summarize = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
```

```python
import gradio as gr

def summarize_text(text):
  return summarize(text, max_length=100, min_length=25)[0]["summary_text"]

def export_summary(summary):
    file_path = "summary.txt"
    with open(file_path, "w") as f:
        f.write(summary)
    # return file_path
    return gr.DownloadButton(label=f"Download file", value=file_path)

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("Text Summarizer with Export Option")

    text_input = gr.Textbox(label="Enter text to summarize", lines=6)
    summarize_btn = gr.Button("Summarize")
    summary_output = gr.Textbox(label="Summary", lines=4)

    export_btn = gr.Button("Export Summary to File")
    download_file = gr.File(label="Download File")

    # Step 1: Summarization
    summarize_btn.click(fn=summarize_text, inputs=text_input, outputs=summary_output)

    # Step 2: Export to file
    export_btn.click(fn=export_summary, inputs=summary_output, outputs=download_file)


demo.launch(share=True)
```
